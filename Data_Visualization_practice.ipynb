```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
diabetes_df=pd.read_csv("/content/diabetes.csv")
```

```python
diabetes_df
```

```python
diabetes_df.describe()
```

```python
sns.countplot(x='Outcome', data=diabetes_df)
plt.title('Distribution of Diabetic vs Non-Diabetic Outcomes');
```

```python
sns.heatmap(diabetes_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Matrix');
```

```python
sns.boxplot(x='Outcome', y='Glucose', data=diabetes_df)
plt.title('Glucose Levels by Diabetes Outcome');
```

```python
sns.histplot(data=diabetes_df, x='BMI')
plt.title('BMI Distribution Overlaid by Outcome');
```

```python
sns.histplot(data=diabetes_df, x='BMI', hue='Outcome')
plt.title('BMI Distribution Overlaid by Outcome');
```

```python
sns.histplot(data=diabetes_df, x='BMI', hue='Outcome', kde=True)
plt.title('BMI Distribution Overlaid by Outcome');
```

```python
sns.histplot(data=diabetes_df, x='BMI', hue='Outcome', kde=True, element='step')
plt.title('BMI Distribution Overlaid by Outcome');
```

```python
sns.histplot(data=diabetes_df, x='BMI', hue='Outcome', kde=True, element='step', stat='density')
plt.title('BMI Distribution Overlaid by Outcome');
```

```python
sns.histplot(data=diabetes_df, x='BMI', hue='Outcome', kde=True, element='step', stat='density', common_norm=False)
plt.title('BMI Distribution Overlaid by Outcome');
```

```python
sns.scatterplot(x='Age', y='Glucose', hue='Outcome', data=diabetes_df, alpha=0.7)
plt.title('Glucose Levels vs Age');
```

```python
sns.pairplot(diabetes_df[['Pregnancies', 'Glucose', 'BMI', 'Age', 'Outcome']], hue='Outcome', corner=True);
```
