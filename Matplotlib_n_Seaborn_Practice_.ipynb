```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import pandas as pd
summary_df=pd.read_csv("/content/online_learning_platform_day 2.csv")
```

```python
summary_df
```

```python
plt.plot(summary_df['Month'], summary_df['Revenue']);
```

```python
plt.figure(figsize=(12,6))
plt.title('Revenue trend by Month')
sns.scatterplot(x=summary_df.Month,y=summary_df.Revenue);
```

```python
sns.barplot(x="Month",y="Course_Completions", data=summary_df)
plt.title('Course Completions by Month')
plt.figure(figsize=(12,6));
```

```python
plt.figure(figsize=(12, 6))
plt.title('Course Completions by Month', fontsize=14, fontweight='bold')
sns.barplot(x="Month", y="Course_Completions", data=summary_df, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.title("Distribution of Watch Time")
plt.hist(summary_df.Avg_Watch_Time);
```

```python
plt.figure(figsize=(12,6))
plt.title('Watch time vs Course Completion')
sns.scatterplot(x=summary_df.Course_Completions,y=summary_df.Avg_Watch_Time,hue=summary_df.Month);
```

```python
summary_pivot=summary_df.pivot_table(index="Month", values=['Active_Users', 'Revenue', 'Course_Completions', 'Avg_Watch_Time'],
                                       aggfunc='mean')
```

```python
summary_pivot
```

```python
sns.heatmap(summary_pivot);
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(summary_df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.bar(summary_df['Month'], summary_df['Revenue'])
plt.figure(figsize=(14,6));
```

```python
plt.figure(figsize=(12, 6))
plt.title('Revenue Trend by Month')
plt.bar(summary_df['Month'], summary_df['Revenue'], color='seagreen')
plt.xticks(rotation=45, ha='right');
```

```python
plt.figure(figsize=(12,6))
plt.title('Correlation between active users and revenue')
sns.scatterplot(x=summary_df.Active_Users,y=summary_df.Revenue);
```
