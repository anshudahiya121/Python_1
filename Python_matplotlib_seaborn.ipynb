```python
setosa_df=flowers_df[flowers_df.species=="setosa"]
versicolor_df=flowers_df[flowers_df.species=="versicolor"]
virginica_df=flowers_df[flowers_df.species=="virginica"]
```

```python
plt.hist(setosa_df.sepal_width)
plt.hist(versicolor_df.sepal_width);
```

```python
plt.hist(setosa_df.sepal_width,alpha=1,bins=np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width,alpha=1,bins=np.arange(2,5,0.25));
```

```python
plt.hist(setosa_df.sepal_width,alpha=1,bins=np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width,alpha=1,bins=np.arange(2,5,0.25))
plt.hist(virginica_df.sepal_width,alpha=1,bins=np.arange(2,5,0.25));
```

```python
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], alpha=1, bins=np.arange(2,5,0.25), stacked=True);
plt.legend(["Setosa","Versicolor","Virginica"]);
```

```python
import seaborn as sns
flights_df=sns.load_dataset("flights")
flights_pivot=flights_df.pivot(index="month", columns="year", values="passengers")
```

```python
flights_pivot
```

```python
sns.heatmap(flights_pivot);
```

```python
sns.heatmap(flights_pivot)
plt.title("Passenger Traffic");
```

```python
import matplotlib.pyplot as plt
sns.heatmap(flights_pivot,fmt='g',annot=True,);
plt.title("Passenger Traffic");
```

```python
sns.heatmap(flights_pivot,fmt='g',annot=True,cmap='YlGnBu',linewidth=0.5,annot_kws={"fontsize":11});
plt.title("Passenger Traffic");
```
