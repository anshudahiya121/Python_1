1.WAp for string array operation str= aman rahul priya simran ..convert all in uppercase , print 1st n last name, find total no. of names

Create an array from 1to24 and reshape it into 6row 4 column ,print first row ,last row,3rd colmn,last 2rows

```python
import numpy as np
arr = np.arange(1, 25).reshape(6, 4)
print(arr)
print(arr[0])
print(arr[-1])
print(arr[:, 2])
print(arr[4:])
```

WAP in array 45,67,89,34,90,56,78,99 find i.)scores above 75 ii.)less then 40 fail iii.)total number of elements in array

```python
import numpy as np
marks=np.array([45,67,89,34,90,56,78,99])
print(len((marks)[marks>75]))
print(len((marks)[marks<40]))
print(len(marks))
```
